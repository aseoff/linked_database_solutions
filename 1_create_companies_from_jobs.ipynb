```python
#inport pandas module
import pandas as pd
```

```python
#making df
jobs100 = pd.read_csv("100.csv")
```

```python
#show preview of jobs
jobs100.head()
```

```python
companies = jobs100["Company"]
companies.head()
```

```python
id = 1
unique_companies = {}
for company in companies:
    if company in unique_companies.values():
        #getID of company
        # list out keys and values separately
        key_list = list(unique_companies.keys())
        val_list = list(unique_companies.values())
        
        # print id with val company
        position = val_list.index(company)
        #print(key_list[position])
    else:
        #setID to next available and add company to unique companies
        unique_companies[id] = company
        #print(unique_companies)
        id = id + 1

#print(unique_companies)
```

```python
#Create new dataframe with company_id and company_name
company_info = pd.DataFrame(columns=["company_id","company_name"])
for key in unique_companies:
    company_info.loc[len(company_info.index)] = [key, unique_companies[key]]

company_info.head()
#company_info.shape
```

```python
#add ids to jobs100 df
jobs100["company_id"] = ""
jobs100.head()
```

```python
for i in range(len(jobs100)):
    company = jobs100.iloc[i].Company
    company_id = company_info.loc[company_info.company_name == company].company_id.values[0]
    jobs100.loc[i, "company_id"] = company_id
jobs100.head()
```

```python
jobs100.to_csv('jobs100.csv')
company_info.to_csv('company_info_100.csv')
```
