```python
#inport pandas module
import pandas as pd
```

```python
#making df
connections_df = pd.read_csv("Connections.csv")
```

```python
#show preview of connections
connections_df.head()
```

```python
#read in company info df
company_info_df = pd.read_csv("company_info_100.csv")
company_info_df.head()
company_info_df.drop(columns=company_info_df.columns[0], axis=1, inplace=True)
company_info_df.head()
```

```python
company_info_df.shape
```

```python
#Create new company id column for connections_df
connections_df["company_id"] = ""
connections_df.head()
```

```python
for i in range(len(connections_df)):
    company = connections_df.iloc[i].Company
    #print(company)
    try:
        company_id = company_info_df.loc[company_info_df.company_name == company].company_id.values[0]
        connections_df.loc[i, "company_id"] = company_id
    except:
        # Need to add a new company to company info and put new id in connections
        new_company_id = len(company_info_df) + 1
        #print(new_company_id)
        connections_df.loc[i, "company_id"] = new_company_id #added new company id to connections table
        new_row = [new_company_id, company]
        company_info_df.loc[len(company_info_df)] = new_row

    
```

```python
connections_df.to_csv("connections100.csv")
company_info_df.to_csv("updated_companies_100.csv")
```
